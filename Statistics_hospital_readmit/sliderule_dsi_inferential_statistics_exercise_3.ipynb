# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## Hypothesis Testing

### Analysis of Variance (ANOVA)
We chose One-Way ANOVA to see if there is a statistically significant difference between groups. In order to get F-statistic, we will need to calculated 
- sum of square within the groups (SSW)
- sum of square between the groups (SSB)
- total sum of square (SST) 
- SST = SSW + SSB

_Null hypothesis(H0): µ0 = µ1 = µ3_ 
_Alternative hypothesis: µ0 ≠ µ1 ≠ µ3_
 
Python Scipy.stats package performs one-way anova that returns F-statistic and P-value.

```python
import scipy.stats as stats
import seaborn as sns
sns.set(color_codes=True)
```

```python
#Divide hospitals/facilities into 3 groups for ANOVA

#Group1 with 'excess readmission ratio' < 100
group1 = clean_hospital_read_df[['Number of Discharges', 'Excess Readmission Ratio']][clean_hospital_read_df['Number of Discharges'] < 100]
#Group1 with 'excess readmission ratio' > 1000
group2 = clean_hospital_read_df[['Number of Discharges', 'Excess Readmission Ratio']][clean_hospital_read_df['Number of Discharges'] > 1000]
#Group1 with 'excess readmission ratio' >= 100 & <=1000
group3 = clean_hospital_read_df[['Number of Discharges', 'Excess Readmission Ratio']][(clean_hospital_read_df['Number of Discharges'] >= 100) & (clean_hospital_read_df['Number of Discharges'] <= 1000)]

#Add column 'group' to each group
group1['group'] = '1'
group2['group'] = '2'
group3['group'] = '3'

#Concatenate group100, group1000 and groupOther
df = pd.concat([group1, group2, group3])
df.isnull().any().any() #Check if the data has nan
df = df.dropna(axis = 0) #Drop all nan
group1 = group1.dropna(axis = 0) #Drop nan in group1
```

#### Check Distribution Normality
Use Python Seaborn distribution plot to see if the data is normally distributed.

```python
#Distribution plot
fig = plt.figure(figsize = (10,6))
sns.distplot(group1['Excess Readmission Ratio'])
plt.show()

fig = plt.figure(figsize = (10,6))
sns.distplot(group2['Excess Readmission Ratio'])
plt.show()

fig = plt.figure(figsize = (10,6))
sns.distplot(group3['Excess Readmission Ratio'])
plt.show()
        
```

```python
#Put three distribution plots together
fig = plt.figure(figsize = (10,6))
sns.distplot(group1['Excess Readmission Ratio'])
sns.distplot(group2['Excess Readmission Ratio'])
sns.distplot(group3['Excess Readmission Ratio'])
plt.show()
```

From the distribution plot, group1 and group3 are skewed to the right, and group2 seems to be normally distributed, but it's hard to tell simply from the distribution plot. Thus, we need to test group2 for normality.

```python
#k is Kurtosis to test if the data is normally distributed. P is p-value.
from scipy import stats

k, p = stats.mstats.normaltest(group3['Excess Readmission Ratio']) 
if p < 0.05:
    print('Not normal distribution')
else:
    print('Normal distribution')
```

#### Boxplot
Use boxplot to see the variance between each group and within each group

```python
#Plot boxplot for 'Excess Readmission Ratio'
fig = plt.figure(figsize = (10,6))
sns.boxplot(x = 'group', y= 'Excess Readmission Ratio', data = df, palette = 'Set3')
plt.show()
```

Although scatter plot from the preliminary analysis can tell us the trend of the data, it doesn't tell us the variance between groups. Compared to scatter plot, boxplot shows us variances between each group and variances within each group. From the boxplot, the variance between group1, group2 and group3 seems small, but the variance within each group is wider than we thought. For example, you can tell from boxplot that group3 has widely spread readmission ratio.

#### Non-parametric ANOVA
Since the data is not normally distributed, we use non-parametric ANOVA test

```python
#ANOVA test

print(stats.kruskal(group1['Excess Readmission Ratio'], group2['Excess Readmission Ratio'], group3['Excess Readmission Ratio']))
```
